```python
def find_similar_shared_nonshared_trips(data_shared, data_nonshared):
    shared_id = []
    non_shared_id = []
    for i in range(0,data_shared.shape[0]):
        shared = data_shared.iloc[i,]
        start_region = shared.start_region
        end_region = shared.end_region
        non_shared = data_nonshared.loc[(data_nonshared.start_region == start_region) & (data_nonshared.end_region == end_region)]
        shared_id.append(shared.order_id)
        non_shared_id.append(find_nearest_trip(shared,non_shared))
    return shared_id,non_shared_id
```

```python
def find_nearest_trip(shared_data,nonshared_data):
    if nonshared_data.shape[0] != 0:
        nearest_trip = nonshared_data.loc[np.argmin((nonshared_data.start_long - shared_data.start_long) ** 2 
                                                + (nonshared_data.end_long - shared_data.end_long) ** 2)]
        nearest_trip_id = nearest_trip.order_id
    else:
        nearest_trip_id = -1
    return nearest_trip_id
```
